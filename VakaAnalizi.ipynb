```python
top_five_artists = spotify.groupby('artist_name').count().sort_values('track_name', ascending=False)[:5]
```

```python
top_five_artists
```

```python
top_five_artists.plot.barh()
plt.show()
```

```python
top_five_loudest_track = spotify.groupby('track_name').sum().sort_values('loudness', ascending=False)[:5]
```

```python
top_five_loudest_track
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='loudness', y = top_five_loudest_track.index, data = top_five_loudest_track)
plt.show()
```

```python
most_danceable_track = spotify[['danceability', 'track_name', 'artist_name']].sort_values('danceability', ascending= False)[:5]
most_danceable_track
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x = 'danceability', y='artist_name', data = most_danceable_track)
plt.show()
```

```python
top_ten_inst_tracks = spotify[['instrumentalness', 'track_name', 'artist_name']].sort_values('instrumentalness', ascending=False)[:10]
top_ten_inst_tracks
```

```python
plt.figure(figsize=(12,5))
plt.pie(x='instrumentalness', data=top_ten_inst_tracks, autopct='%1.2f%%')
plt.title('Top 10 INST Tracks')
plt.show()
```

```python
pop_more_than_90 = spotify[spotify['popularity']>90].sort_values('popularity',ascending=False)
pop_more_than_90.head()
```

```python
sns.displot(pop_more_than_90['acousticness'], color = 'green')
plt.show()
# Akustik olma durumunun populer olma durumuyla bir baglantisi olmadigini goruyoruz
# Tersk korelasyon oldgunu goruyoruz ne kadar akustikse o kadar az populer
```
