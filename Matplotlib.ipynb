### Bir grafikte 2 cizgi gosterme

```python
plt.figure(figsize = (10, 5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x, 2)
y2 = np.power(x, 3)
plt.plot(x, y1, 'b-', label = '$y1 = x^2$')
plt.plot(x, y2, 'r-', label = '$y2 = x^3$' )
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend() #plot da yazdigimiz labellari eklemek icin
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 2000)
plt.figure(figsize=(10,6))
plt.plot(x, np.sin(x), label= '$Sin(X)$')
plt.plot(x, np.cos(x), label= '$Cos(X)$')
plt.xlabel(r'$X$', fontsize= 18) # Labellari buyulttuk
plt.ylabel(r'$Y$', fontsize= 18)
plt.title('$ Sin(x) $ $ & $ $ Cos(x) $', fontsize= 14) # Baslik verdik
plt.legend(loc = 'upper right') # Labeli sag uste al dedik
plt.show()
```

```python
plt.figure(figsize = (10, 5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x, 3)
plt.plot(x, y1,'b-', label = '$y1=x^2$')
plt.plot(x, y2, color='red', linewidth = 1.0, linestyle = '--' )
plt.xlabel('X-Axis',fontsize = 14)
plt.ylabel('Y-Axis', fontsize = 14)
plt.legend(loc = 'upper center', fontsize='large')
plt.show()
```

### 4 cizgili grafik

```python
plt.figure(figsize = (16,9))
x = np.linspace(0, 10, 2000)
plt.plot(x,  np.sin(x),label = '$Sin(X) Dashed$', linestyle='dashed')
plt.plot(x+1, np.sin(x), label = '$Sin(X) Dashdot$', linestyle='dashdot')
plt.plot(x, np.cos(x), label = '$Cos(X) Solid$', linestyle='solid')
plt.plot(x+1, np.cos(x), label = '$Cos(X) Dotted$', linestyle='dotted')
plt.xlabel(r'$X$', fontsize=14)
plt.ylabel(r'$Y$', fontsize=14)
plt.legend(loc = 'upper right', fontsize = 14, bbox_to_anchor =(1.2, 1.0))
plt.show()
```
