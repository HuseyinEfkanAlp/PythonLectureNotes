### Bir grafikte 2 cizgi gosterme

```python
plt.figure(figsize = (10, 5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x, 2)
y2 = np.power(x, 3)
plt.plot(x, y1, 'b-', label = '$y1 = x^2$')
plt.plot(x, y2, 'r-', label = '$y2 = x^3$' )
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend() #plot da yazdigimiz labellari eklemek icin
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(0, 10, 2000)
plt.figure(figsize=(10,6))
plt.plot(x, np.sin(x), label= '$Sin(X)$')
plt.plot(x, np.cos(x), label= '$Cos(X)$')
plt.xlabel(r'$X$', fontsize= 18) # Labellari buyulttuk
plt.ylabel(r'$Y$', fontsize= 18)
plt.title('$ Sin(x) $ $ & $ $ Cos(x) $', fontsize= 14) # Baslik verdik
plt.legend(loc = 'upper right') # Labeli sag uste al dedik
plt.show()
```

```python
plt.figure(figsize = (10, 5))
x = np.array([1,2,3,4,5,6])
y1 = np.power(x,2)
y2 = np.power(x, 3)
plt.plot(x, y1,'b-', label = '$y1=x^2$')
plt.plot(x, y2, color='red', linewidth = 1.0, linestyle = '--' )
plt.xlabel('X-Axis',fontsize = 14)
plt.ylabel('Y-Axis', fontsize = 14)
plt.legend(loc = 'upper center', fontsize='large')
plt.show()
```

```python
plt.figure(figsize = (8,8))
area = [48, 30, 20, 15]
total = np.sum(area)
labels = ['Low', 'Medium', 'High', 'Critical']

def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)

colors = ['#8BC34A', '#D4E157', '#FFB300', '#FF7043']
plt.pie(area, labels = labels, colors = colors, startangle = 45, shadow = True, autopct = val_per)
plt.show()
```

```python
sira = [1,3,5]
boy = [180, 170, 165]
kilo = [60,90,60]
ayakkabi = [43,36,39]
sinif = [3,4,2]

fig = plt.figure(figsize=(12,10))
spec = fig.add_gridspec(2,3, hspace=0.25, wspace=0.25)

ax = fig.add_subplot(spec[0,:])
ax.plot(sira,boy, marker='o',linestyle='--')
ax.set_title('Boy Grafigi')
ax.set_xlabel('Sira')
ax.set_ylabel('Boy')
ax = fig.add_subplot(spec[1,0])
ax.plot(sira,kilo, marker='v', linestyle='--')
ax.set_title('Kilo Grafigi')
ax.set_xlabel('Sira')
ax.set_ylabel('Kilo')
ax = fig.add_subplot(spec[1,1])
ax.plot(sira,ayakkabi,marker = '^', color = 'red')
ax.set_title('Ayakkabi Grafigi')
ax.set_xlabel('Sira')
ax.set_ylabel('Ayakkabi')
ax = fig.add_subplot(spec[1,2])
ax.plot(sira,sinif,marker = 'X', color = 'red')
ax.set_title('Sinif Grafigi')
ax.set_xlabel('Sira')
ax.set_ylabel('Sinif')
plt.show()
```
